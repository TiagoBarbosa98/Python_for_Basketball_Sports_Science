```python
import pandas as pd
import numpy as np
```

```python
# üèÄ Load Sample Raw Basketball Data
data = {
    "Player": ["LeBron James", "Anthony Davis", "Austin Reaves", "D'Angelo Russell", "LeBron James"],
    "PTS": [27, 22, np.nan, 18, 27],
    "REB": [8, 11, 4, None, 8],
    "AST": [7, 3, 5, 6, 7],
    "MIN": [35, 34, 30, 32, None],
    "Position": ["F", "F/C", "G", "G", "F"]
}
df = pd.DataFrame(data)
print("Raw Data:")
display(df)
```

```python
# üßπ 1Ô∏è‚É£ Remove Duplicates
df = df.drop_duplicates()
print("\nAfter Removing Duplicates:")
display(df)
```

```python
# üß≠ 2Ô∏è‚É£ Identify Missing Values
print("\nMissing Values Summary:")
print(df.isna().sum())
```

```python
# üß© 3Ô∏è‚É£ Fill Missing Values
df["PTS"].fillna(df["PTS"].mean(), inplace=True)
df["REB"].fillna(df["REB"].median(), inplace=True)
df["MIN"].fillna(df["MIN"].mean(), inplace=True)

print("\nAfter Filling Missing Values:")
display(df)
```

```python
# üß† 4Ô∏è‚É£ Fix Data Types
df["MIN"] = df["MIN"].astype(int)
```

```python
# üè∑Ô∏è 5Ô∏è‚É£ Rename Columns
df.rename(columns={"PTS": "Points", "REB": "Rebounds", "AST": "Assists"}, inplace=True)
```

```python
# üßæ 6Ô∏è‚É£ Reorder Columns
df = df[["Player", "Position", "Points", "Rebounds", "Assists", "MIN"]]

print("\nCleaned Basketball Dataset:")
display(df)
```

```python
# üíæ 7Ô∏è‚É£ Save Clean Dataset
df.to_csv("clean_basketball_data.csv", index=False)
print("\n‚úÖ Cleaned dataset saved as 'clean_basketball_data.csv'")
```
