```python
import pandas as pd
import numpy as np
```

```python
# ðŸ“‚ Step 1: Create or Load a Raw Dataset
data = {
    "Player": ["Player A", "Player B", "Player C", "Player D", "Player E"],
    "PTS": [25, None, 18, 30, 12],
    "REB": [7, 9, np.nan, 10, 5],
    "AST": [5, 6, 4, None, 3],
    "MIN": [33, 28, 30, 40, np.nan],
    "Position": ["G", "F", "G", "C", "G"]
}
df = pd.DataFrame(data)
print("Raw Dataset:")
display(df)
```

```python
# ðŸ§¹ Step 2: Check for Missing Values
print("\nMissing Values Summary:")
print(df.isnull().sum())
```

```python
# ðŸ§© Step 3: Fill or Drop Missing Data (your choice)
# Example:
df["PTS"].fillna(df["PTS"].mean(), inplace=True)
df["AST"].fillna(df["AST"].median(), inplace=True)
df["REB"].fillna(df["REB"].mean(), inplace=True)
df["MIN"].fillna(df["MIN"].mean(), inplace=True)
```

```python
# âš™ï¸ Step 4: Fix Data Types
df["MIN"] = df["MIN"].astype(int)
```

```python
# ðŸ·ï¸ Step 5: Rename Columns
df.rename(columns={"PTS": "Points", "REB": "Rebounds", "AST": "Assists"}, inplace=True)
```

```python
# ðŸ§­ Step 6: Detect Outliers (Optional)
# Example: Flag players with minutes > 45
df["Outlier_MIN"] = df["MIN"] > 45
```

```python
# âœ… Step 7: Export Clean Data
print("\nCleaned Dataset:")
display(df)
```

```python
# ðŸ’¾ Save to CSV
df.to_csv("cleaned_week7_assignment.csv", index=False)
print("\nâœ… Cleaned dataset saved as 'cleaned_week7_assignment.csv'")
```
