## Instructions

2. Collect the following stats:
- Field Goal Attempts (FGA)
- Free Throw Attempts (FTA)
- Offensive Rebounds (OREB)
- Turnovers (TO)
- Points Scored (PTS)

3. Write a Python script to:
- Estimate possessions using the formula:  
  `Possessions = FGA + (0.44 × FTA) – OREB + TO`
- Calculate Offensive Efficiency:  
  `(PTS / Possessions) × 100`

4. Print the results in a clear sentence.

```python
# Step 1: Enter game stats here
# 11/11/2025 LAL @ CHA

FGA = 90   # Field Goal Attempts
FTA = 20   # Free Throw Attempts
OREB = 12  # Offensive Rebounds
TO = 11    # Turnovers
PTS = 121   # Points Scored
```

```python
# Step 2: Calculate possessions and offensive efficiency
possessions = FGA + (0.44 * FTA) - OREB + TO
off_efficiency = (PTS / possessions) * 100 if possessions > 0 else 0
print("Lakers scored", PTS, "points on", possessions, "possessions.")
print("Offensive Efficiency:", off_efficiency)
```
