```python
import pandas as pd

# ðŸ€ Create a Simple Box Score DataFrame
data = {
    "Player": ["LeBron James", "Anthony Davis", "Austin Reaves", "D'Angelo Russell"],
    "PTS": [27, 22, 15, 19],
    "REB": [8, 11, 4, 3],
    "AST": [7, 3, 5, 6]
}

df = pd.DataFrame(data)

# Display the initial DataFrame
df
```

```python
# ðŸ§© Explore the Data
print("Shape:", df.shape)
print("\nColumns:", df.columns)
print("\nSummary Stats:")
print(df.describe())
```

```python
# ðŸ” Select Columns & Filter Rows
print("Points column (Series):\n", df["PTS"])
print("\nPlayers with >20 PTS:\n", df[df["PTS"] > 20])
```

```python
# ðŸ§® Add and Modify Columns
# Efficiency (EFF) is often calculated as PTS + REB + AST
df["EFF"] = df["PTS"] + df["REB"] + df["AST"]

# Display the DataFrame with the new EFF column
df
```

```python
# Sorting by Efficiency
df_sorted = df.sort_values(by="EFF", ascending=False)

# Display the sorted DataFrame
df_sorted
```

```python
# ðŸ’¾ Exporting Data
# index=False prevents writing the DataFrame's index as a column in the CSV
df_sorted.to_csv("week4_team_boxscore.csv", index=False)
print("Data saved as week4_team_boxscore.csv")
```
