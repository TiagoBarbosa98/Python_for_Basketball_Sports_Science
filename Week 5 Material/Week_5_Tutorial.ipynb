# Week 5 ‚Äì Data Visualization with Matplotlib & Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
# üèÄ Create Sample Data
data = {
    "Player": ["LeBron James", "Anthony Davis", "Austin Reaves", "D'Angelo Russell"],
    "PTS": [27, 22, 15, 19],
    "REB": [8, 11, 4, 3],
    "AST": [7, 3, 5, 6]
}
df = pd.DataFrame(data)
df
```

```python
# üìä Bar Plot ‚Äì Player Points
plt.figure(figsize=(8, 5))
sns.barplot(x="Player", y="PTS", data=df, palette="coolwarm")
plt.title("Points per Game by Player")
plt.ylabel("Points")
plt.xticks(rotation=15)
plt.show()
```

```python
# ‚öæ Scatter Plot ‚Äì Rebounds vs Assists
plt.figure(figsize=(6, 5))
sns.scatterplot(x="REB", y="AST", data=df, s=100, color="teal")
plt.title("Rebounds vs Assists")
plt.xlabel("Rebounds")
plt.ylabel("Assists")
plt.show()
```

```python
# üìà Line Plot ‚Äì Simulated Points Over Games
games = ["G1", "G2", "G3", "G4", "G5"]
points = [20, 25, 28, 22, 30]
```

```python
plt.figure(figsize=(7, 4))
plt.plot(games, points, marker="o", color="orange")
plt.title("LeBron James ‚Äì Points Over 5 Games")
plt.xlabel("Game")
plt.ylabel("Points")
plt.show()
```
