# Week 2 – Control Flow & Functions
Welcome to Week 2! This week we’ll cover conditionals, loops, and functions using basketball examples.

## Section A: Control Flow – If/Else

```
points = 112

if points > 110:
    print("High scoring game")
elif points >= 90:
    print("Average scoring game")
else:
    print("Low scoring game")
```

## Section B: Loops

```
player_points = [25, 18, 12, 15, 10]

total_points = 0
for p in player_points:
    total_points += p

print("Team total points:", total_points)
```

## Section C: Functions

```
def effective_fg(FGM, threePM, FGA):
    return (FGM + 0.5 * threePM) / FGA

efg = effective_fg(10, 3, 20)
print("Effective FG%:", round(efg, 3))
```

## Section D: Multi-Stat Function

```
def advanced_stats(FGM, FGA, threePM, FTM, FTA, PTS):
    eFG = (FGM + 0.5 * threePM) / FGA
    TS = PTS / (2 * (FGA + 0.44 * FTA))
    return {"eFG%": round(eFG, 3), "TS%": round(TS, 3)}

stats = advanced_stats(10, 20, 3, 6, 8, 29)
print(stats)
```
