## Step 1: Define Functions

```python
# Define a function to calculate eFG%
def effective_fg(FGM, threePM, FGA):
    return (FGM + 0.5*threePM) / FGA

# Define a function to calculate TS%
def true_shooting(PTS, FGA, FTA):
    return PTS / (2*(0.44*FTA + FGA))

```

## Step 2: Input Player Stats and Calculate

```python
# Example stats: Player line
FGM = 10
FGA = 20
threePM = 3
FTM = 6
FTA = 8
PTS = 29

# Call your functions here and print the results
# Example:
print("Effective FG%:", round(effective_fg(FGM, threePM, FGA), 3))
print("True Shooting %:", round(true_shooting(PTS, FGA, FTA), 3))
```

## Step 3: Efficiency Message

```python
# Example if/else statement
efg = 0.575  # Replace with your function output

if efg > 0.6:
    print("Elite shooting night!")
elif efg >= 0.5:
    print("Above average efficiency.")
else:
    print("Below average efficiency.")
```

## Extra Challenge: Multiple Players

```python
# Example: List of players with box score lines
players = [
    {"name": "Player A", "FGM": 10, "FGA": 20, "3PM": 3, "FTM": 6, "FTA": 8, "PTS": 29},
    {"name": "Player B", "FGM": 5, "FGA": 12, "3PM": 1, "FTM": 2, "FTA": 2, "PTS": 13},
    {"name": "Player C", "FGM": 8, "FGA": 15, "3PM": 2, "FTM": 4, "FTA": 5, "PTS": 22},
]

# Loop through players, calculate stats, and print results
for player in players:
    efg = effective_fg(FGM=player["FGM"], threePM=player["3PM"], FGA=player["FGA"])
    player['eFG'] = efg

print("Most efficient player is ", max(players, key=lambda x: x["eFG"])["name"], "with eFG%:", max(players, key=lambda x: x["eFG"])["eFG"])


```
