```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# üèÄ Step 1: Create Your Dataset
players = np.array(["Player A", "Player B", "Player C", "Player D", "Player E"])
points = np.array([25, 19, 22, 15, 10])
rebounds = np.array([9, 6, 11, 8, 5])
assists = np.array([7, 5, 6, 4, 3])
```

```python
# üßÆ Step 2: Calculate Key Metrics
avg_points = np.mean(points)
std_points = np.std(points)
median_rebounds = np.median(rebounds)
corr_pr = np.corrcoef(points, rebounds)[0, 1]
```

```python
print("Average Points:", avg_points)
print("Std Dev Points:", std_points)
print("Median Rebounds:", median_rebounds)
print("Correlation (Points vs Rebounds):", round(corr_pr, 2))
```

```python
# ‚ö° Step 3: Compute Player Efficiency
efficiency = points * 0.5 + rebounds * 1.2 + assists * 0.8
df = pd.DataFrame({
    "Player": players,
    "Points": points,
    "Rebounds": rebounds,
    "Assists": assists,
    "Efficiency": efficiency
})
df
```

```python
# üìä Step 4: Visualize Your Findings
plt.figure(figsize=(8,5))
plt.bar(df["Player"], df["Efficiency"], color="purple")
plt.title("Player Efficiency Comparison")
plt.ylabel("Efficiency Score")
plt.show()
```
