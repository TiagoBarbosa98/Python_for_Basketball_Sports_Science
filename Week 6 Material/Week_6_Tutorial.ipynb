```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# üèÄ Create Sample Basketball Data
players = np.array(["LeBron James", "Anthony Davis", "Austin Reaves", "D'Angelo Russell", "Jarred Vanderbilt"])
points = np.array([27, 22, 15, 18, 8])
rebounds = np.array([8, 11, 4, 3, 6])
assists = np.array([7, 3, 5, 6, 2])
```

```python
# üìä Descriptive Statistics
print("Average Points:", np.mean(points))
print("Median Rebounds:", np.median(rebounds))
print("Standard Deviation (Assists):", np.std(assists))
```

```python
# ‚ö° Broadcasting Example
efficiency = points * 0.5 + rebounds * 1.2 + assists * 0.8
for i in range(len(players)):
    print(f"{players[i]} Efficiency: {efficiency[i]:.2f}")
```

```python
# üìà Correlation Example
corr_points_reb = np.corrcoef(points, rebounds)[0, 1]
print("Correlation (Points vs Rebounds):", round(corr_points_reb, 2))
```

```python
# üßÆ Team Summary DataFrame
df = pd.DataFrame({
    "Player": players,
    "Points": points,
    "Rebounds": rebounds,
    "Assists": assists,
    "Efficiency": efficiency
})
df
```

```python
# üìä Visualization
plt.figure(figsize=(8,5))
plt.scatter(points, rebounds, s=100, color="teal")
for i, name in enumerate(players):
    plt.text(points[i]+0.3, rebounds[i], name, fontsize=9)
plt.title("Points vs Rebounds")
plt.xlabel("Points")
plt.ylabel("Rebounds")
plt.show()
```
